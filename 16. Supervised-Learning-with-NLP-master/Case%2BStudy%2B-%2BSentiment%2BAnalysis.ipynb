### Data Prep

```python
import pandas as pd
import numpy as np

# Read in the data
df = pd.read_csv('Amazon_Unlocked_Mobile.csv')


# Sample the data to speed up computation
# Comment out this line to match with lecture
df = df.sample(frac=0.1, random_state=10)

df
```

```python
df.isnull().sum()
```

```python
# Drop missing values
df.dropna(inplace=True)

# Remove any 'neutral' ratings equal to 3
df = df[df['Rating'] != 3]

# Encode 4s and 5s as 1 (rated positively)
# Encode 1s and 2s as 0 (rated poorly)
df['Positively_Rated'] = np.where(df['Rating'] > 3, 1, 0)
df.head(10)
df.shape
```

# CountVectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer

# Fit the CountVectorizer to the training data
vect = CountVectorizer().fit(X_train)
```

```python
vect
```

```python
vect.get_feature_names()[::2000]
```

```python
len(vect.get_feature_names())
```

```python
# transform the documents in the training data to a document-term matrix
X_train_vectorized = vect.transform(X_train)
df_vect = pd.DataFrame(X_train_vectorized.toarray(), columns=vect.get_feature_names())

```

```python
df_vect
```

```python
from sklearn.linear_model import LogisticRegression

# Train the model
model = LogisticRegression()
model.fit(X_train_vectorized, y_train)
```

```python
from sklearn.metrics import roc_auc_score

# Predict the transformed test documents
predictions = model.predict(vect.transform(X_test))

print('AUC: ', roc_auc_score(y_test, predictions))
```

```python
# get the feature names as numpy array
feature_names = np.array(vect.get_feature_names())

# Sort the coefficients from the model
sorted_coef_index = model.coef_[0].argsort()

# Find the 10 smallest and 10 largest coefficients
# The 10 largest coefficients are being indexed using [:-11:-1] 
# so the list returned is in order of largest to smallest
print('Smallest Coefs:\n{}\n'.format(feature_names[sorted_coef_index[:10]]))
print('Largest Coefs: \n{}'.format(feature_names[sorted_coef_index[:-11:-1]]))
```

# Tfidf

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Fit the TfidfVectorizer to the training data specifiying a minimum document frequency of 5
vect_tfidf = TfidfVectorizer(min_df=5).fit(X_train)
len(vect.get_feature_names())
```

```python
X_train_vectorized = vect_tfidf.transform(X_train)

model = LogisticRegression()
model.fit(X_train_vectorized, y_train)

predictions = model.predict(vect_tfidf.transform(X_test))

print('AUC: ', roc_auc_score(y_test, predictions))
```

# n-grams

```python
# Fit the CountVectorizer to the training data specifiying a minimum 
# document frequency of 5 and extracting 1-grams and 2-grams
vect = CountVectorizer(min_df=5, ngram_range=(1,2)).fit(X_train)

X_train_vectorized = vect.transform(X_train)

len(vect.get_feature_names())
```

```python
model = LogisticRegression()
model.fit(X_train_vectorized, y_train)

predictions = model.predict(vect.transform(X_test))

print('AUC: ', roc_auc_score(y_test, predictions))
```

```python
feature_names = np.array(vect.get_feature_names())

sorted_coef_index = model.coef_[0].argsort()

print('Smallest Coefs:\n{}\n'.format(feature_names[sorted_coef_index[:10]]))
print('Largest Coefs: \n{}'.format(feature_names[sorted_coef_index[:-11:-1]]))
```

```python
# These reviews are now correctly identified
print(model.predict(vect.transform(['not an issue, phone is working',
                                    'an issue, phone is not working'])))
```

# Dimensionality Reduction with LDA

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

data = pd.read_csv('reviews_sentiments.csv')



cv = CountVectorizer(stop_words = 'english', max_features=250)
cv_ft = cv.fit_transform(data.review_text)
lda = LinearDiscriminantAnalysis()
lda.fit(cv_ft.toarray(), data.sentiment)
coef_df = pd.DataFrame({'word':cv.get_feature_names(),'coef':lda.coef_[0]}).sort_values(by='coef',ascending=False).reset_index(drop=True)
#print(coef_df.head())
```
