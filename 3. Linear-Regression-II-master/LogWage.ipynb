```python
#import libraries
# Y = logwage where logwage values have no outliers.
import pandas as pd

import matplotlib.pyplot as plt
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.compat import lzip
import statsmodels.stats.api as sm
from sklearn.metrics import mean_squared_error
from math import sqrt
from sklearn.model_selection import train_test_split
import statsmodels as sm
import statsmodels.formula.api as smf
from sklearn import linear_model

%matplotlib inline 
# keeps plots within notebook
```

```python
df = pd.read_csv('Wage.csv', index_col=0, skiprows=0)
df.shape
df.head()
#encoder = LabelEncoder() 
#encoder.fit(data['education']) encoder.classes_ 
#df['education'] = encoder.transform(df['education'])


df['education'] = df['education'].map( {'1. < HS Grad':1, '2. HS Grad':2, '3. Some College':3, '4. College Grad':4, '5. Advanced Degree':5} )
df['health'] = df['health'].map( {'1. <=Good':1, '2. >=Very Good':2} )
df['jobclass'] = df['jobclass'].map( {'1. Industrial':1, '2. Information':2} )
df1 = df[df.columns[0:2]] 
df2 = df[df.columns[5:6]]
df3 = df[df.columns[7: 9]]
df4 = pd.concat([df1, df2, df3], axis=1)
#df4['Age**2'] = df['age']*df['age']
df4.head()
#print(df4)
target = pd.DataFrame(df, columns=["logwage"])
#print(target)
df4.head()


```

OLS Assumption1

```python
print("Rows,Columns:",df4.shape)
df4.describe()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(df4, target, test_size=0.30)
#print(X_train)
```

```python
# Fitting using stats model
model_1 = smf.OLS(Y_train,X_train).fit()
prediction_1 = model_1.predict(X_train) # predict y_train values
prediction_2 = model_1.predict(X_test)# predict y_test values
#w_predict = model_1.predict(X_train)
#residual = df["wage"].values - w_predict
#print(residual)
model_1.summary()
```

```python
rmse_training = sqrt(mean_squared_error(Y_train, prediction_1))
print("RMSE for training data:", rmse_training)
rmse_test = sqrt(mean_squared_error(Y_test,prediction_2))
print("RMSE for test data:", rmse_test)
```

```python
vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(X_test.values, i) for i in range(X_test.shape[1])]
vif["features"] = X_test.columns
vif
```
